```python
import numpy as np
import sys
from matplotlib import pyplot as plt
from CellularAutomata import CellularAutomata
np.set_printoptions(threshold=sys.maxsize)
```

We create the model with `CellularAutomata` object. We define `L` as grid width and `rule` as a cellular automata rule number. 

Additionally we can define various `boundary_conditions`:
* periodic
* reflective
* constant
* null
* random

Other string in input will raise an error.

```python
L = 30
boundary_condition = 'periodic'
rule = 30
```

```python
cellular_automata = CellularAutomata(L=L, boundary_conditions=boundary_condition)
```

We can assign a rule to automata object. It is then converted into a dictionary of all possible options: 

```python
cellular_automata.change_rule(rule)
cellular_automata.cases
```

```python
cellular_automata.initialize_simulation(random=True)
print(cellular_automata.get_grid())
```

```python
cellular_automata.reset_simulation()
cellular_automata.initialize_simulation(random=False)
print(cellular_automata.get_grid())
```

```python
cellular_automata.simulation_step()
print(cellular_automata.get_grid())
```

Let's simulate for 20 steps and see how it affects the 2D grid.

```python
cellular_automata.simulation(20)
print(cellular_automata.get_grid())
```

Let's visualize what we have achieved.

```python
figure1, axes1 = plt.subplots(1, 1, layout='constrained')
cellular_automata.plot_grid(axes1, add_title=False)
```

Now we can check for bigger values for rules 18, 26, 181 and 61

```python
L = 200
steps = 400
boundary_condition = 'periodic'
cellular_automata = CellularAutomata(L=L, boundary_conditions=boundary_condition)
```

```python
figure2, axes2 = plt.subplots(1, 3, layout='constrained', figsize=(10, 8))
for rule, axes in zip([18, 26, 181], figure2.axes):
    cellular_automata.change_rule(rule)
    cellular_automata.reset_simulation()
    cellular_automata.initialize_simulation(random=False)
    cellular_automata.simulation(steps)
    cellular_automata.plot_grid(axes)
    axes.set_title(f'Rule {rule}')
figure2.suptitle(f'Evolution of Cellular Automata for {steps} steps, L = {L}')
figure2.savefig('./images/cellular_automata.png')
```

```python
rule = 57
L = 200
steps = 300
boundary_conditions = ['periodic', 'reflective', 'constant', 'opposite']
figure3, axes3 = plt.subplots(2, 2, layout='constrained', figsize=(8, 10))
cellular_automata.change_rule(rule)
cellular_automata.reset_simulation()
cellular_automata.change_grid_size(L)
for condition, axes in zip(boundary_conditions, figure3.axes):
    cellular_automata.change_boundary_conditions(condition)
    cellular_automata.reset_simulation()
    cellular_automata.initialize_simulation(random=False)
    cellular_automata.simulation(steps)
    cellular_automata.plot_grid(axes)
    axes.set_title(f'{condition}')
figure3.suptitle(f'Evolution of Cellular Automata for different boundary conditions and L = {L}')
figure3.savefig('./images/automata_boundary_conditions.png')
```

```python
rule = 30
L = 200
steps = 300
boundary_conditions = ['periodic', 'reflective', 'constant', 'opposite']
figure3, axes3 = plt.subplots(2, 2, layout='constrained', figsize=(8, 10))
cellular_automata.change_rule(rule)
for condition, axes in zip(boundary_conditions, figure3.axes):
    cellular_automata.change_boundary_conditions(condition)
    cellular_automata.reset_simulation()
    cellular_automata.initialize_simulation(random=True)
    cellular_automata.simulation(steps)
    cellular_automata.plot_grid(axes)
    axes.set_title(f'{condition}')
figure3.suptitle(f'Evolution of Cellular Automata rule {rule}, L = {L}\n'
                 f'Different boundary conditions and random input')
figure3.savefig(f'./images/cellular_automata_boundary_rule{rule}.png')
```

```python
rule = 61
L = 500
steps = 1000
boundary_condition = 'periodic'
cellular_automata.reset_simulation()
cellular_automata.change_rule(rule)
cellular_automata.change_grid_size(L)
cellular_automata.change_boundary_conditions(boundary_conditions=boundary_condition)
cellular_automata.initialize_simulation()
cellular_automata.simulation(steps)
```
